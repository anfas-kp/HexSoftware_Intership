## Load Iris Dataset



```python
import seaborn as sns

iris_df = sns.load_dataset('iris')

print("Iris dataset loaded successfully. First 5 rows:")
iris_df.head()
```

## Explore Dataset Structure


Display the first few rows, column names, and data types to understand the dataset's structure.


```python
print("First 5 rows of the Iris dataset:")
print(iris_df.head())

print("\nColumn names of the Iris dataset:")
print(iris_df.columns)

print("\nData types and non-null counts of the Iris dataset:")
iris_df.info()
```

## Check for Missing Values


Identify and count any missing values across all columns in the dataset.


```python
print("Missing values in each column of the Iris dataset:")
print(iris_df.isnull().sum())
```

## Visualize Feature Distributions



### Subtask: Display Correlation Matrix

**Reasoning:**
To understand the linear relationships between the numerical features in the `iris_df`, I will calculate the correlation matrix. A heatmap visualization of this matrix will provide a clear and intuitive representation of how each feature correlates with the others.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix for numerical columns
correlation_matrix = iris_df.corr(numeric_only=True)

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Iris Features')
plt.show()
```

### Subtask: Visualize Feature Distributions using Pairplot

**Reasoning:**
To further visualize the relationships between all numerical features, and also the distribution of each feature segmented by the `species` column, I will use a pairplot. A pairplot creates a grid of pairwise relationships between variables in a dataset, with distributions of each variable shown on the diagonal. This will help us understand both individual feature distributions and correlations between features, particularly how they vary across different Iris species.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pairplot to visualize relationships between numerical features, colored by 'species'
plt.figure(figsize=(10, 8))
sns.pairplot(iris_df, hue='species', diag_kind='kde')
plt.suptitle('Pairplot of Iris Features by Species', y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get numerical columns
num_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Create histograms for each numerical feature
for col in num_cols:
    plt.figure(figsize=(8, 6))
    plt.hist(iris_df[col], bins=10, edgecolor='black')
    plt.title(f'Distribution of {col.replace("_", " ").title()}')
    plt.xlabel(f'{col.replace("_", " ").title()}')
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

### Summary of Exploration and Visualizations

**Dataset Structure:**
*   The Iris dataset contains 150 entries and 5 columns: `sepal_length`, `sepal_width`, `petal_length`, `petal_width` (all `float64`), and `species` (`object`).
*   The first few rows show consistent, well-formed data.

**Missing Values:**
*   No missing values were found across any of the columns, confirming the dataset's cleanliness.

**Feature Distributions (Histograms):**
*   **sepal_length:** The distribution appears somewhat normal, with a peak around 5.0-6.0 cm.
*   **sepal_width:** This feature also shows a somewhat normal distribution, centered roughly around 3.0-3.5 cm.
*   **petal_length:** The distribution is clearly bimodal or multimodal, with a distinct peak at lower values (around 1.0-2.0 cm) and another broader peak at higher values (around 4.0-5.0 cm). This suggests the presence of distinct groups within the data, likely corresponding to different species.
*   **petal_width:** Similar to petal_length, petal_width also exhibits a bimodal or multimodal distribution, with one peak at very low values (around 0.0-0.5 cm) and another at higher values (around 1.0-2.0 cm). This further reinforces the idea of distinct groups.

**Overall Insight:**
The bimodal/multimodal distributions observed in `petal_length` and `petal_width` are strong indicators that these features are highly discriminative between the different species of Iris flowers. The `species` column, being an object type, is the target variable that these numerical features likely help distinguish.
